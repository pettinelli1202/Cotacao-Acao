```python
import yfinance
import matplotlib.pyplot as plt
import pyautogui
import pyperclip
import time
```

```python
ticker = input("Digite a sigla da ação desejada: ")

plt.title('Movimentação durante seis meses')
plt.ylabel('Cotação da ação em R$')
plt.xlabel('Meses')
dados = yfinance.Ticker(ticker).history("6mo")
plt.suptitle(ticker.upper())
fechamento = dados.Close
fechamento.plot()
```

```python
cota_max = fechamento.max()
cota_min = fechamento.min()
atual = fechamento[-1]
```
